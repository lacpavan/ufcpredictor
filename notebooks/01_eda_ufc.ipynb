# UFC Winner Predictor - Exploratory Data Analysis

This notebook documents the first analysis steps behind the UFC winner prediction project.

Goals:
- inspect the processed fight dataset
- review target balance
- check missing values and numeric distributions
- support the feature engineering and model decisions used in the project


```python
from pathlib import Path

import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

sns.set_theme(style='whitegrid')

PROJECT_ROOT = Path.cwd().resolve().parent if Path.cwd().name == 'notebooks' else Path.cwd().resolve()
DATA_PATH = PROJECT_ROOT / 'data' / 'raw' / 'ufc_fights.csv'

df = pd.read_csv(DATA_PATH, parse_dates=['fight_date'])
df.head()
```

```python
print('Rows:', len(df))
print('Columns:', len(df.columns))
print('\nColumns:')
print(df.columns.tolist())
```

```python
missing = df.isna().sum().sort_values(ascending=False)
missing[missing > 0].head(20)
```

```python
winner_share = df['winner'].value_counts(dropna=False)
winner_share
```

```python
plt.figure(figsize=(6, 4))
sns.countplot(data=df, x='winner', order=df['winner'].value_counts().index, palette=['#d62828', '#3a86ff'])
plt.title('Winner distribution')
plt.xlabel('Corner')
plt.ylabel('Fight count')
plt.show()
```

```python
numeric_cols = [
    'red_age', 'blue_age', 'red_height_cm', 'blue_height_cm',
    'red_reach_cm', 'blue_reach_cm', 'red_wins', 'blue_wins',
    'red_losses', 'blue_losses', 'red_sig_str_acc', 'blue_sig_str_acc',
    'red_td_acc', 'blue_td_acc'
]

df[numeric_cols].describe().T
```

```python
plot_cols = ['red_age', 'blue_age', 'red_wins', 'blue_wins', 'red_sig_str_acc', 'blue_sig_str_acc']
df[plot_cols].hist(figsize=(12, 8), bins=25)
plt.suptitle('Selected feature distributions')
plt.tight_layout()
plt.show()
```

## Notes

- The current repository uses a temporal split instead of a random split to better match the real prediction setting.
- The current production model is a `HistGradientBoostingClassifier` from `scikit-learn`.
- SHAP-based interpretability is planned as a next step and is not yet implemented in the current version.
